### Feature Engineering

```python
df['year']=df['Order Date'].dt.year
```

```python
df['month']=df['Order Date'].dt.month_name()
```

```python
df['Delivery Days']=(df['Ship Date']-df['Order Date']).dt.days
```

```python
df['Profit Margin']=np.round((df['Profit']/df['Sales'])*100,2)
```
